```python
import os
import csv
import pandas as pd
import numpy as np
```

```python
load_file = "astronauts.csv"

astronauts_df = pd.read_csv(load_file)
```

```python
astronauts_df = astronauts_df.dropna(subset=['Missions'])
# astronauts_df = astronauts_df[astronauts_df['Year'].between(1998,2009)]
astronauts_df_split = astronauts_df["Name"].str.split(" ")
astronauts_df_name = [astronauts_df_split[-1] for astronauts_df_split in astronauts_df_split]
astronauts_df["Last Name"] = astronauts_df_name
astronauts_df = astronauts_df.dropna(subset=['Last Name'])
astronauts_df
```

```python
# astronauts_df
degree_count = astronauts_df.groupby("Undergraduate Major")
degree_count = degree_count["Name"].count()

undergrade_degree_df = pd.DataFrame(degree_count)
sort = undergrade_degree_df.sort_values("Name", ascending=False)
sort.head(10)
```

```python
grad_degree_count = astronauts_df.groupby("Graduate Major")
grad_degree_count = grad_degree_count["Name"].count()

grad_degree_df = pd.DataFrame(grad_degree_count)
sort1 = grad_degree_df.sort_values("Name", ascending=False)
sort1.head(10)
```
